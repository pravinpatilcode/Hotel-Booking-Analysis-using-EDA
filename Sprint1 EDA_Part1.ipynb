```python
def sort_data(df):   #defining a fun to sort the data
    cat = []
    num = []
    for i in df.columns:
        if df[i].nunique()< 15:  #if the data contatin less than 15 unique values , i will be a categorical
            cat.append(i)   #appending to cat
        else:
            num.append(i)   #aopending to num
    return cat, num
```

```python
def desc_num(col):
    print("Missing values in", col, "are", df[col].isnull().sum())
    print("Mean:", df[col].mean() ,"    Median:",df[col].median() , "   Mode:", df[col].mode()[0])
    print("Range:", df[col].max()- df[col].min())
    q1= df[col].quantile(0.25)
    q3= df[col].quantile(0.75)
    iqr= q3-q1
    print("IQR:", iqr)
    print('Standard Deviation:', df[col].std())
    lf= q1-1.5*iqr
    uf= q3+1.5*iqr
    condition_outlier= (df[col]<lf)| (df[col]>uf)
    print("lower fence:",lf)
    print("Upper fence:",uf)
    print("Total Outliers:", condition_outlier.sum())
```

```python
desc_num('Age')
```

```python
def vis_num(col):
    fig, ax= plt.subplots(1,2, figsize= (10,6))
    sns.histplot(df[col], ax= ax[0], color= 'm')
    ax[0].set_title("histogram for"+col)
    sns.boxplot(df[col], ax= ax[1], color= 'g')
    ax[1].set_title("boxplot for"+col)

    plt.suptitle("Analysis of "+col)
    plt.show
```

```python
vis_num('Age')
```

```python
#Interpretation : By considering Histogram and boxplot the data is normally distributed with no outliers in the AGE column.
```
