```python
def sort_data(df):
    cat = []
    num = []
    for i in df.columns:
        if df[i].dtype == 'object':
            cat.append(i)
        else:
            num.append(i)
    return cat, num
```

# TASK 3

```python
filter1 = df[df['BusinessTravel'] == 'Travel_Rarely']
filter1
```

```python

def cat_cat(cat1, cat2, data):
    # Count occurrences
    counts = data.groupby([cat1, cat2]).size().reset_index(name='count')
    
    # Pivot the data to get categories on the y-axis and counts for each hue category
    pivot_table = counts.pivot(index=cat1, columns=cat2, values='count').fillna(0)
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Plot horizontal bars for each hue category
    pivot_table.plot(kind='barh', stacked=False, ax=ax, colormap='tab20')
    
    # Customize the plot
    plt.xlabel('Count')
    plt.ylabel(cat1)
    plt.title(" Job role wise count of attrited employees who travels Rarely")
    plt.legend(title=cat2)
    plt.show()
```

```python
cat_cat('JobRole', 'Attrition', filter1)
```
